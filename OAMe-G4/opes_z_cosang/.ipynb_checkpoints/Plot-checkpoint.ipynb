```python
import scipy.interpolate

N = 50 #number of points for plotting/interpolation

m = 21   #number of contours

levels = np.linspace(0,40,m)

x, y, z = np.genfromtxt(r'fes_2d', unpack=True)

xi = np.linspace(x.min(), x.max(), N)
yi = np.linspace(y.min(), y.max(), N)
zi = scipy.interpolate.griddata((x, y), z, (xi[None,:], yi[:,None]), method='cubic')

X, Y = np.meshgrid(xi, yi)

fig = plt.figure()
cp = plt.contourf(X, Y, zi, m, levels=levels, cmap='coolwarm')
#plt.clim(0.0,55.0)
cbar = plt.colorbar(cp)
cbar.set_label('FES (kJ/mol)',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Deep LDA CV",fontsize=16)
plt.ylabel("z (nm)",fontsize=16)

trajlist = [ 'trace_126_40.xtc','trace_294_29.xtc','trace_363_47.xtc','trace_376_38.xtc','trace_377_41.xtc','trace_504_42.xtc','trace_567_36.xtc','trace_568_34.xtc','trace_575_39.xtc','trace_840_40.xtc' ]

for traj in trajlist:
    trjname = traj[:-4]
    #print(trjname)
    #os.system('gmx_mpi trjconv -s topol.tpr -f %s.xtc -pbc mol -o %s_mol.xtc'%(trjname,trjname))
    #os.system('plumed driver --plumed plumed_descriptors.dat --ixtc %s_mol.xtc'%trjname)
    #os.system('mv COLVAR_d COLVAR_%s'%trjname)

    data = np.loadtxt('../westpa_mab_3/merged_trajectories/COLVAR_%s'%trjname)

    deep_lda_model = torch.jit.load('modelG4_a.pt')

    x_data = torch.Tensor(data[:,2:])
    with torch.no_grad():
        water_lda = deep_lda_model(x_data)
    water_lda = water_lda.detach().numpy()

    dcv = water_lda
    plt.scatter(dcv[::5],data[::5,1],s=0.2,color='k',alpha=1)



plt.savefig('projection_on_2d.png',dpi=300,bbox_inches='tight')
```
